**4.What are probability distribution functions(PDF)?**

Ans: A **Probability Distribution Function (PDF)** describes how probabilities are distributed over the possible values of a random variable.

- For **discrete random variables**, it's called a **probability mass function (PMF)**, which gives the probability for each specific value.
- For **continuous random variables**, it's called a **probability density function (PDF)**, which gives the relative likelihood of the variable taking a particular value (though the probability of any exact value is 0, it can be calculated over a range).

**5.How do cumulative distribution functions(CDF) differ from probability distribution functions(PDF)?**

Ans: The key difference between **Cumulative Distribution Function (CDF)** and **Probability Distribution Function (PDF)** is:

1. **PDF (Probability Distribution Function)**:
   - Describes the probability of a random variable taking a specific value.
   - For discrete variables, it's called **PMF (Probability Mass Function)**.
   - For continuous variables, it shows the likelihood of a value within an interval but doesn't give exact probabilities for single values (since they are 0).

2. **CDF (Cumulative Distribution Function)**:
   - Describes the probability that a random variable is **less than or equal to** a certain value.
   - It gives the cumulative probability up to a certain point.
   - The CDF is the **accumulation** of the PDF, starting from the lowest value up to the given point.

**7.What are the key properties of a Bernoulli distribution?**

1. **Two outcomes**: The random variable can only take values 0 (failure) or 1 (success).
2. **Single trial**: It models a single trial or experiment (like flipping a coin once).

**8.What is the binomial distribution and how is it used in probability?**

Ans: The **binomial distribution** is a probability distribution that models the number of successes in a fixed number of **independent trials**, where each trial has two possible outcomes (success or failure).

1. **Fixed number of trials (n)**: The experiment is repeated a set number of times (e.g., flipping a coin 10 times).
2. **Probability of success (p)**: Each trial has the same probability of success (e.g., getting heads on a coin flip).
3. **Outcomes**: It counts the number of successes (like how many heads you get in 10 coin flips).

**Formula:**
The probability of getting exactly **k** successes in **n** trials is given by the binomial formula:

\[
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
\]

Where:
- \( n \) = number of trials
- \( k \) = number of successes
- \( p \) = probability of success on each trial



**9.What is the Poisson distribution and where is it applied?**

Ans: The **Poisson distribution** is a probability distribution that models the number of **events** happening in a fixed interval of time or space, when the events occur **randomly** and **independently**.

**Key features:**
1. **Fixed interval**: The number of events is counted in a specific time period or space (e.g., number of cars passing a street in an hour).
2. **Constant average rate (λ)**: Events happen at a constant average rate (e.g., on average, 5 customers arrive per minute).
3. **Rare events**: It’s often used when events are rare and happen randomly.

**Formula:**
The probability of exactly **k** events happening in a time interval is:

\[
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
\]

Where:
- \( \lambda \) = average number of events in the interval
- \( k \) = number of events you're calculating the probability for

**Applications:**
- **Number of emails received in an hour**.
- **Car accidents happening on a road**.
- **Number of phone calls at a call center**.

**11.What are the characteristics of a normal distribution?**

2. **Mean = Median = Mode**: The highest point of the curve is at the mean, and it’s also the median and mode.
3. **Symmetry**: It is symmetrical around the mean, meaning the left side is a mirror image of the right side.
4. **Defined by two parameters**:
   - **Mean (μ)**: The center of the distribution (where the peak is).
   - **Standard deviation (σ)**: Measures the spread of the distribution; smaller σ means the curve is narrow, larger σ means it's wider.
5. **68-95-99.7 Rule**:
   - About 68% of the data lies within 1 standard deviation of the mean.
   - About 95% lies within 2 standard deviations.
   - About 99.7% lies within 3 standard deviations.

**13.What is the Central Limit Theorem(CLT),and why is it critical in statistics?**

Ans: The **Central Limit Theorem (CLT)** states that, no matter the shape of the original population distribution, the **sampling distribution of the sample mean** will be approximately **normal** if the sample size is large enough.


- As the sample size increases, the distribution of sample means gets closer to a normal distribution.
- This holds true even if the population data itself isn’t normally distributed.
- **Critical for statistics** because it allows us to make inferences about a population using the normal distribution, even with non-normal data.


The CLT shows that with a large enough sample, the average of your samples will follow a normal distribution, which is important for making reliable statistical conclusions.

**15.What is the application of Z statistics in hypothesis testing?**

1. **Standardize**: The Z-score is calculated to see how many standard deviations the sample mean is away from the population mean.
2. **Compare**: The Z-score is compared to a critical value from the standard normal distribution to decide whether to **reject** or **fail to reject** the null hypothesis.

Z statistics help measure how unusual or extreme a sample result is, and they are used in hypothesis testing to make decisions about the population based on sample data.

**16.How do you calculate a Z-score,and what does it represent?**

Ans: To calculate a **Z-score**, use this formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:
- \( X \) = the value you're analyzing (e.g., a sample mean or data point).
- \( \mu \) = the population mean.
- \( \sigma \) = the population standard deviation.


The **Z-score** tells you how many **standard deviations** the value \( X \) is away from the population mean.
- A **Z-score of 0** means \( X \) is exactly at the mean.
- A **positive Z-score** means \( X \) is above the mean.
- A **negative Z-score** means \( X \) is below the mean.


A Z-score measures how far a data point is from the mean, in terms of standard deviations.

**17.What are point estimates and interval estimates in statistics?**

Ans: In statistics:

1. **Point Estimate**:
   - A **single value** used to estimate a population parameter (like the mean or proportion).
   - Example: If the average height of a sample is 5.6 feet, then 5.6 feet is the point estimate of the population mean.

2. **Interval Estimate**:
   - A **range of values** that likely contains the population parameter, with a certain level of confidence.
   - Example: A 95% confidence interval for the population mean might be between 5.4 and 5.8 feet, meaning we're 95% confident the true population mean lies within that range.

- **Point estimate** gives a single best guess.
- **Interval estimate** gives a range with a confidence level.

- The **confidence level** (e.g., 95%) tells you how often the interval will contain the true value if you repeat the study many times.

Confidence intervals help us understand the **uncertainty** of our estimates and give a range where the true value is likely to fall.

**6.Write a Python program to calculate and plot the cumulative distribution function(CDF) of a discrete uniform distributions.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of possible outcomes for the discrete uniform distribution
outcomes = np.arange(1, 7)  # For a fair die, outcomes are 1, 2, 3, 4, 5, 6

# Define the probability for each outcome (all probabilities are equal in a uniform distribution)
probabilities = np.ones_like(outcomes) / len(outcomes)

# Calculate the CDF by accumulating the probabilities
cdf = np.cumsum(probabilities)

# Plot the CDF
plt.step(outcomes, cdf, where='post', color='blue', label='CDF', linewidth=2)
plt.xlabel('Outcomes')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of a Discrete Uniform Distribution')
plt.xticks(outcomes)
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True)
plt.show()
```

**10.Implement the Central Limit Theorem(CTL) using Python for a non-normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the exponential distribution
lambda_ = 1  # Rate parameter for the exponential distribution
sample_size = 30  # Size of each sample
num_samples = 1000  # Number of samples to draw

# Generate data from an exponential distribution (non-normal)
data = np.random.exponential(scale=1/lambda_, size=10000)

# List to store the means of each sample
sample_means = []

# Take multiple samples and compute their means
for _ in range(num_samples):
    sample = np.random.choice(data, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot the original non-normal data
plt.figure(figsize=(12, 6))

# Plot the histogram of the original non-normal distribution
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, color='skyblue', alpha=0.7)
plt.title('Original Non-Normal Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot the histogram of the sample means (CLT in action)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, color='skyblue', alpha=0.7)
plt.title('Sampling Distribution of the Sample Means (CLT)')
plt.xlabel('Mean of Samples')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

**15.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 0      # Mean of the distribution
std_dev = 1   # Standard deviation of the distribution
sample_size = 30  # Size of each sample
num_samples = 1000  # Number of samples to draw

# Generate multiple samples from the normal distribution
sample_means = []

for _ in range(num_samples):
    sample = np.random.normal(mean, std_dev, sample_size)
    sample_means.append(np.mean(sample))

# Plot the histogram of the sample means
plt.figure(figsize=(10, 6))

# Plot the original distribution (normal distribution)
plt.subplot(1, 2, 1)
data = np.random.normal(mean, std_dev, 10000)
plt.hist(data, bins=30, density=True, color='skyblue', alpha=0.7)
plt.title('Original Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot the sampling distribution of the sample means (CLT)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, color='skyblue', alpha=0.7)
plt.title('Sampling Distribution of the Sample Means (CLT)')
plt.xlabel('Mean of Samples')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

Explanation:
np.random.normal(mean, std_dev, sample_size): This generates random values from a normal distribution with a specified mean and standard deviation.
Sampling: We draw 1,000 samples, each with a size of 30, from the normal distribution and compute their means.
CLT Verification: According to the Central Limit Theorem, the distribution of the sample means should approximate a normal distribution, even though we are sampling from a normal distribution in the first place.
Plotting: We create two histograms:
The first histogram shows the original normal distribution.
The second histogram shows the sampling distribution of the sample means, which should also appear as a normal distribution.

Result:
The first histogram will show the original normal distribution with a bell shape.
The second histogram will show the sampling distribution of the sample means. As the sample size increases and we take more samples, the distribution of the means will look increasingly like a normal distribution, confirming the Central Limit Theorem.
Since we're starting with a normal distribution, the sampling distribution of the sample means will already be approximately normal, but this approach still demonstrates the general concept of CLT for different distributions.

**16.Write a Python function to calculate and plot the standard normal distribution(mean=0, std=1).**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    # Parameters for the standard normal distribution
    mean = 0
    std_dev = 1

    # Generate values for the x-axis (range of values)
    x = np.linspace(-4, 4, 1000)  # Values from -4 to 4 (covering most of the distribution)

    # Calculate the PDF for the standard normal distribution
    pdf_values = norm.pdf(x, mean, std_dev)

    # Plot the standard normal distribution
    plt.plot(x, pdf_values, label="Standard Normal Distribution", color='blue', linewidth=2)
    plt.fill_between(x, pdf_values, color='skyblue', alpha=0.4)

    # Add labels and title
    plt.title('Standard Normal Distribution (mean=0, std=1)')
    plt.xlabel('X')
    plt.ylabel('Probability Density')

    # Display the plot
    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function to plot the standard normal distribution
plot_standard_normal_distribution()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-score for the given data point x
    z_score = (x - mean) / std_dev

    return z_score, mean, std_dev

def plot_standard_normal_distribution():
    # Generate a range of values for the standard normal distribution
    x_values = np.linspace(-4, 4, 1000)
    y_values = norm.pdf(x_values, 0, 1)  # PDF of standard normal distribution

    # Plot the standard normal distribution
    plt.plot(x_values, y_values, label="Standard Normal Distribution", color='blue', linewidth=2)
    plt.fill_between(x_values, y_values, color='skyblue', alpha=0.4)
    plt.axvline(x=0, color='black', linestyle='--')  # Line at the mean (0)
    plt.title('Standard Normal Distribution (mean=0, std=1)')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.legend()

def main():
    # Example dataset
    data = np.random.normal(50, 10, 1000)  # Generate 1000 data points with mean 50 and std 10

    # Data point to calculate Z-score for
    x = 60  # Example data point

    # Calculate the Z-score for the data point
    z_score, mean, std_dev = calculate_z_score(data, x)

    print(f"Data Point: {x}")
    print(f"Mean of Data: {mean:.2f}")
    print(f"Standard Deviation of Data: {std_dev:.2f}")
    print(f"Z-score for {x}: {z_score:.2f}")

    # Plot the standard normal distribution and the Z-score
    plot_standard_normal_distribution()

    # Highlight the Z-score on the plot
    plt.axvline(x=z_score, color='red', linestyle='--', label=f"Z-score = {z_score:.2f}")
    plt.legend()

    # Show the plot
    plt.show()

# Run the program
main()
```

**19.Implement hypothesis testing using Z- statistics for a simple dataset.**

```python
import numpy as np
import scipy.stats as stats

def z_test(sample, hypothesized_mean, population_std_dev, alpha=0.05):
    # Sample size
    n = len(sample)

    # Calculate the sample mean
    sample_mean = np.mean(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - hypothesized_mean) / (population_std_dev / np.sqrt(n))

    # Calculate the critical value for a two-tailed test
    critical_value = stats.norm.ppf(1 - alpha/2)

    # Print the results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"Critical Value: ±{critical_value:.2f}")

    # Hypothesis testing decision
    if abs(z_statistic) > critical_value:
        print("Reject the Null Hypothesis (H0).")
    else:
        print("Fail to Reject the Null Hypothesis (H0).")

# Example dataset
sample_data = np.array([45, 48, 50, 47, 49, 51, 46, 48, 50, 52])

# Parameters for hypothesis test
hypothesized_mean = 50  # Hypothesized population mean
population_std_dev = 5  # Known population standard deviation
alpha = 0.05  # Significance level (5%)

# Perform the Z-test
z_test(sample_data, hypothesized_mean, population_std_dev, alpha)
```

**20.Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    # Calculate sample mean and sample standard deviation
    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    n = len(data)  # Sample size

    # Calculate the Z-critical value for the given confidence level
    alpha = 1 - confidence_level
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Calculate the margin of error
    margin_of_error = z_critical * (sample_std_dev / np.sqrt(n))

    # Calculate the confidence interval
    confidence_interval_lower = sample_mean - margin_of_error
    confidence_interval_upper = sample_mean + margin_of_error

    return sample_mean, confidence_interval_lower, confidence_interval_upper

# Example dataset
data = [45, 48, 50, 47, 49, 51, 46, 48, 50, 52]

# Calculate confidence interval
sample_mean, lower, upper = confidence_interval(data, confidence_level=0.95)

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

# Interpretation
if lower <= sample_mean <= upper:
    print(f"The sample mean lies within the 95% confidence interval: ({lower:.2f}, {upper:.2f})")
else:
    print(f"The sample mean is outside the 95% confidence interval.")
```

**22.Write a Python Script to calculate and visualize the probability density function(PDF) of a normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev):
    # Generate a range of x values (from mean - 4*std_dev to mean + 4*std_dev)
    x_values = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

    # Calculate the PDF of the normal distribution for each x value
    pdf_values = norm.pdf(x_values, mean, std_dev)

    # Plot the PDF
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, pdf_values, label=f'Normal Distribution (mean={mean}, std={std_dev})', color='blue', linewidth=2)
    plt.fill_between(x_values, pdf_values, alpha=0.2, color='skyblue')  # Shade the area under the curve

    # Add titles and labels
    plt.title('Probability Density Function (PDF) of Normal Distribution', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.grid(True)
    plt.legend(loc='upper right')

    # Show the plot
    plt.show()

# Parameters for the normal distribution
mean = 0      # Mean of the distribution
std_dev = 1   # Standard deviation of the distribution

# Call the function to plot the PDF of the normal distribution
plot_normal_pdf(mean, std_dev)
```

**23.Use Python to calculate and interpret the cumulative distribution function(CDF) of a Poisson distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam, max_value):
    # Generate x values (0 to max_value) for which we calculate the CDF
    x_values = np.arange(0, max_value + 1)

    # Calculate the CDF of the Poisson distribution
    cdf_values = poisson.cdf(x_values, lam)

    # Plot the CDF
    plt.figure(figsize=(10, 6))
    plt.step(x_values, cdf_values, label=f'Poisson CDF (λ={lam})', where='post', color='blue', linewidth=2)
    plt.fill_between(x_values, cdf_values, alpha=0.2, color='skyblue')  # Shade the area under the curve

    # Add titles and labels
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (λ={lam})', fontsize=14)
    plt.xlabel('Number of Events (k)', fontsize=12)
    plt.ylabel('Cumulative Probability', fontsize=12)
    plt.grid(True)
    plt.legend(loc='upper left')

    # Show the plot
    plt.show()

# Parameters for the Poisson distribution
lambda_value = 5  # Average rate of occurrence (λ)
max_k = 15  # Maximum value of k to plot

# Call the function to plot the CDF of the Poisson distribution
plot_poisson_cdf(lambda_value, max_k)

# Calculate and print the CDF at specific points
x = np.arange(0, max_k + 1)
cdf_at_points = poisson.cdf(x, lambda_value)

for k, cdf_val in zip(x, cdf_at_points):
    print(f'P(X ≤ {k}) = {cdf_val:.4f}')
```

**28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.**

Ans: To calculate the skewness and kurtosis of a dataset, we can use the scipy.stats module in Python, which provides functions to compute these statistics.

1. Skewness:
Skewness measures the asymmetry of the data distribution.
A skewness of 0 indicates a symmetric distribution.
Positive skew indicates a longer right tail, and negative skew indicates a longer left tail.
2. Kurtosis:
Kurtosis measures the "tailedness" of the distribution.
A kurtosis of 3 indicates a normal distribution.
Values greater than 3 indicate a distribution with heavier tails (leptokurtic).
Values less than 3 indicate lighter tails (platykurtic).

```python
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

# Function to calculate skewness and kurtosis
def calculate_skewness_kurtosis(data):
    skewness_value = skew(data)
    kurtosis_value = kurtosis(data)
    return skewness_value, kurtosis_value

# Generate dataset from normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate skewness and kurtosis
skewness_value, kurtosis_value = calculate_skewness_kurtosis(data)

# Print the results
print(f"Skewness: {skewness_value:.2f}")
print(f"Kurtosis: {kurtosis_value:.2f}")

# Plot the dataset
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Dataset')
plt.show()
```
